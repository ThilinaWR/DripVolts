For the beam vibration equation shown:



The governing equation for transverse vibration of an Euler--Bernoulli beam with damping, axial load, and external forcing is:

\begin{equation}
\rho A \frac{\partial^2 y(x,t)}{\partial t^2}
+ c \frac{\partial y(x,t)}{\partial t}
+ EI \frac{\partial^4 y(x,t)}{\partial x^4}
+ P \frac{\partial^2 y(x,t)}{\partial x^2}
= q(x,t)
\end{equation}

### Where

- $y(x,t)$: Transverse deflection of the beam  
- $x$: Spatial coordinate along the beam length  
- $t$: Time  
- $E$: Young's modulus  
- $I$: Second moment of area  
- $\rho$: Material density  
- $A$: Cross-sectional area  
- $c$: Damping coefficient  
- $P$: Axial load  
- $q(x,t)$: External distributed load
